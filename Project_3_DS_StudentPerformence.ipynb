```python
paid_oh = pd.get_dummies(df["paid"], prefix= 'paid')
higher_oh = pd.get_dummies(df["higher"], prefix= 'higher')
Pstatus_oh = pd.get_dummies(df["Pstatus"], prefix= 'Pstatus')
internet_oh = pd.get_dummies(df['internet'], prefix= 'internet')

df2 = pd.concat([df, Pstatus_oh, paid_oh, higher_oh, internet_oh], axis = 1)

df2 = df2.drop(['paid', 'higher', 'Pstatus', 'internet'], axis = 1)
df2.head()
```
